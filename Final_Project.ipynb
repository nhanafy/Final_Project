### <p style="font-family: Fantasy, fantasy; font-size: 40px; letter-spacing: 5px; color: #C2B5FA; ">What is Ethereum?</p>
Ethereum is a decentralized blockchain platform that establishes a peer-to-peer network that securely executes and verifies application code, called smart contracts. Smart contracts allow participants to transact with each other without a trusted central authority, founded by Vitalik Buterin and Gavin Wood in 2015, today Ethereumâ€™s market capitalization represents more than 17% of the $1.2 trillion global crypto market.
Ether (ETH), the native cryptocurrency of the Ethereum network, is the second most popular digital token after bitcoin (BTC)

### <p style="font-family: Fantasy, fantasy; font-size: 40px; letter-spacing: 5px; color: #C2B5FA; ">What is ERC20?</p>
Ethereum Request for Comment 20 (ERC-20) is the implemented standard for fungible tokens created using the Ethereum blockchain. ERC-20 guides the creation of new tokens on the Ethereum blockchain so that they are interchangeable with other smart contract tokens.
ERC20 is the standard protocol for creating Ethereum-based tokens, which can be utilized and deployed in the Ethereum network.

```python
for i in X_train.columns:
    X_train[i] = pd.to_numeric(X_train[i], errors='coerce')
```

```python
# Define Min Max Scaler and fit transform on X_train
MMscaler = MinMaxScaler(feature_range = (0,1))
X_train_MMscaler= MMscaler.fit_transform(X_train)
print(X_train_MMscaler)

# Define Standard Scaler and fit transform on X_train
stan_scaler = StandardScaler()
X_train_Sscaler = stan_scaler.fit_transform(X_train)
print(X_train_Sscaler)
```

```python
# Scaling using Min Max Scaling

X_train_MMscaler_df = pd.DataFrame(X_train_MMscaler, columns = X_train.columns)
print('Min Max Scaler X_train Dataframe: ',X_train_MMscaler_df.head(3))

X_train_Sscaler_df = pd.DataFrame(X_train_Sscaler, columns = X_train.columns)
print('Standard Scaler X_train Dataframe: ', X_train_Sscaler_df.head(3))
```

<p style="padding:10px;background-color:#5E5A80;margin:0;color:white;font-family:newtimeroman;font-size:150%;text-align:center;border-radius: 15px 50px;overflow:hidden;font-weight:500">For the Imbalance Use the Oversampling Technique (SMOTE)</p>

```python
oversample = SMOTE()
```

```python

X_train_MM_new, y_train_MM_new = oversample.fit_resample(X_train_MMscaler_df, y_train)
print(X_train_MM_new.shape, y_train_MM_new.shape)

X_train_S_new, y_train_S_new = oversample.fit_resample(X_train_Sscaler_df, y_train)
print(X_train_S_new.shape, y_train_S_new.shape)
```

```python
# Target distribution before SMOTE
non_fraud = 0
fraud = 0

for i in y_train:
    if i == 0:
        non_fraud +=1
    else:
        fraud +=1

# Target distribution after SMOTE
smote_non_fraud = 0
smote_fraud = 0

for i in y_train_S_new:
    if i == 0:
        smote_non_fraud +=1
    else:
        smote_fraud +=1


print(f'BEFORE OVERSAMPLING \n \tNon-frauds: {non_fraud} \n \tFauds: {fraud}')
print(f'AFTER OVERSAMPLING \n \tNon-frauds: {smote_non_fraud} \n \tFauds: {smote_fraud}')
```

### <p style="padding:10px;background-color:#5E5A80;margin:0;color:white;font-family:newtimeroman;font-size:150%;text-align:center;border-radius: 15px 50px;overflow:hidden;font-weight:500">Apply & Test Models</p>

**Logistic Regression**

```python
LR = LogisticRegression(random_state=42,max_iter=150)
LR.fit(X_train_S_new, y_train_S_new)

# Apply Standard Scaler on test data
Sscaler_test = stan_scaler.transform(X_test)
X_test_Sscaler_df = pd.DataFrame(Sscaler_test, columns = X_test.columns)
pred_LR = LR.predict(X_test_Sscaler_df) 

print(sklm.classification_report(y_test, pred_LR))
#print(sklm.confusion_matrix(y_test, pred_LR))
#sklm.plot_confusion_matrix(LR, X_test_Sscaler_df, y_test)
cm = sklm.confusion_matrix(y_test, pred_LR, labels=LR.classes_)
disp = sklm.ConfusionMatrixDisplay(confusion_matrix=cm ,
                              display_labels=LR.classes_)
disp.plot()
```

**<p>XGBoost Classifier</p>**

```python
XGB = xgb.XGBClassifier(random_state=42)

XGB.fit(X_train_S_new, y_train_S_new)

pred_XGB = XGB.predict(X_test_Sscaler_df)

print(sklm.classification_report(y_test, pred_XGB))
#print(sklm.confusion_matrix(y_test, pred_XGB))
#sklm.plot_confusion_matrix(XGB, X_test_Sscaler_df, y_test)
cm = sklm.confusion_matrix(y_test, pred_XGB)
disp = sklm.ConfusionMatrixDisplay(confusion_matrix=cm ,
                              display_labels=XGB.classes_)
disp.plot()
```

**<p>Random Forest Classifier</p>**

```python
RF = RandomForestClassifier(random_state=42)

RF.fit(X_train_S_new, y_train_S_new)

pred_RF = RF.predict(X_test_Sscaler_df)

print(sklm.classification_report(y_test, pred_RF))
#print(sklm.confusion_matrix(y_test, pred_RF))

cm = sklm.confusion_matrix(y_test, pred_RF)
disp = sklm.ConfusionMatrixDisplay(confusion_matrix=cm ,
                              display_labels=RF.classes_)
disp.plot()
```

<p> We can see that the best model is XGBoost Classifier </p>
<p> Let's see if we can improve the results more

**Tuning for XGB Classifier**

```python
params_grid = {'learning_rate':[0.01, 0.1, 0.5],
              'n_estimators':[100,200],
              'subsample':[0.5, 0.9],
               'max_depth':[3,4],
               'colsample_bytree':[0.3,0.7]}

grid = GridSearchCV(estimator=XGB, param_grid=params_grid, scoring='recall', cv = 10, verbose = 0)

grid.fit(X_train_S_new, y_train_S_new)
print(f'Best params found for XGBoost are: {grid.best_params_}')
print(f'Best recall obtained by the best params: {grid.best_score_}')
```

```python

preds_best_xgb = grid.best_estimator_.predict(X_test_Sscaler_df)
print(sklm.classification_report(y_test, preds_best_xgb))
# print(sklm.confusion_matrix(y_test, preds_best_xgb))

cm = sklm.confusion_matrix(y_test, preds_best_xgb)
disp = sklm.ConfusionMatrixDisplay(confusion_matrix=cm ,
                              display_labels= grid.classes_)
disp.plot()
```
